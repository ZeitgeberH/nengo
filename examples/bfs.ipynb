```python
import numpy as np
np.random.seed(124)

import matplotlib.pyplot as plt

import nengo
import nengo.spa as spa
from nengo.utils.compat import OrderedDict
from nengo.utils.distributions import Uniform, UniformHypersphere
from nengo.networks.associative import AutoAssociative, HeteroAssociative
```

```python
from collections import defaultdict

class DFA(object):
    
    def __init__(self, states, transitions, initial):
        self.states = states
        self.transitions = transitions
        self.initial = initial
        
    @property
    def adjacency_dict(self):
        adj_list = defaultdict(list)
        for pre, post in self.transitions:
            adj_list[pre].append(post)
        return adj_list
    
    @property
    def actions(self, state_module='state', state_weight=0.5, rule_weight=0.5, 
                state_gain=5.0, utility_offset=-1.0):
        action_matches = {}
        action_effects = {}
        
        # TODO: automatically do 1 - rest for None transitions
        
        for (pre, post), utility in self.transitions.iteritems():
            if utility is None:
                utility = '1.0'
            for i, subutility in enumerate(utility.split(' or ')):
                transition_name = '%s_TO_%s_%d' % (pre, post, i)
                action_matches[transition_name] = '%s*dot(%s, F_%s) + %s*(%s) + %s' % (
                    state_weight, state_module, pre, rule_weight, 
                    subutility, utility_offset)
                action_effects[transition_name] = '%s=%s*%s' % (
                    state_module, state_gain, post)
        
        for post, effect in self.states.iteritems():
            latch_matches = []
            latch_effects = []
            all_effects = effect.split(',')
            for subeffect in all_effects:
                # Every module in the state's effect must have _latched
                lvalue, rvalue = subeffect.split('=', 1)
                latch = lvalue.strip()
                latch_matches.append('dot(%s_latched, 1)' % latch)
                latch_effects.append('%s_latch=1' % latch)
                
            action_matches[post] = '%s*dot(%s, %s) + %s' % (
                state_weight, state_module, post, 1 + utility_offset)
            action_effects[post] = '%s, %s, %s=%s' % (
                effect, ', '.join(latch_effects), state_module, post)
                
            action_matches['F_%s' % post] = '%s*dot(%s, %s) + %s + %s' % (
                state_weight, state_module, post, 
                ' + '.join(latch_matches), 
                1.5 - len(all_effects) + utility_offset)
            action_effects['F_%s' % post] = '%s=%s*F_%s' % (
                state_module, state_gain, post)
    
        actions = {}
        for name in action_matches:
            actions[name] = '%s  -->  %s' % (
                action_matches[name], action_effects[name])
        
        return actions
```

```python
class BFS(spa.SPA):
    
    control_dfa = DFA(
        states = {
            'INIT': 'graph_key=v_0, queue_head=v_0, queue_tail=v_0, queue_learning=-1',
            'IF': 'queue_key=graph_edge',
            'IN_Q': 'temp=graph_next',
            'NEXT_ADJ': 'graph_key=temp',
            'NOTIN_Q': 'queue_key=queue_tail, queue_value=graph_edge, queue_learning=1',
            'ADDED_Q': 'queue_tail=graph_edge, queue_learning=-1, temp=graph_next',
            'END_ADJ': 'queue_key=queue_head',
            'POP_Q': 'queue_head=queue_output, graph_key=queue_output',
        }, 
        transitions = {
            ('INIT', 'IF') : None,
            ('IF', 'NOTIN_Q') : '1.0 - dot(queue_has_key, 1) - dot(queue_key, queue_tail)',
            ('IF', 'IN_Q') : 'dot(queue_has_key, 1) or dot(queue_key, queue_tail)',
            ('IN_Q', 'NEXT_ADJ') : None,
            ('NOTIN_Q', 'NOTIN_Q') : '0.2 + dot(queue_error, queue_error)',
            ('NOTIN_Q', 'ADDED_Q') : '0.5 - dot(queue_error, queue_error)',
            ('ADDED_Q', 'NEXT_ADJ') : None,
            ('NEXT_ADJ', 'IF') : '0.5 - dot(graph_key, NULL)',
            ('NEXT_ADJ', 'END_ADJ') : '-0.5 + dot(graph_key, NULL)',
            ('END_ADJ', 'POP_Q') : None,
            ('POP_Q', 'IF') : None,
        },
        initial = 'INIT')

    def __init__(self, adjacency_dict, initial_vertex, dimensions_graph, dimensions_state, neurons_per_item=500):
        super(BFS, self).__init__()
        
        # Put the graph into a static Dict
        initial_edge_keys = []
        initial_edge_values = []
        initial_next_keys = []
        initial_next_values = []
        v_0 = 'V_%s' % initial_vertex
        num_vertices = len(adjacency_dict)
        graph_vocab = self.get_default_vocab(dimensions_graph)
        for i, adj in adjacency_dict.items():
            for num, j in enumerate(adj):
                key = 'V_%s' % i if num == 0 else 'P_%s_%d' % (i, num)
                next_key = 'P_%s_%d' % (i, num + 1) if num != len(adj) - 1 else 'NULL' 
                initial_edge_keys.append(graph_vocab.parse(key).v)
                initial_edge_values.append(graph_vocab.parse('V_%s' % j).v)
                initial_next_keys.append(graph_vocab.parse(key).v)
                initial_next_values.append(graph_vocab.parse(next_key).v)
        
        max_capacity = AutoAssociative.calculate_max_capacity(dimensions_graph, 0.5)
        
        self.graph_key = spa.DoubleLatch(dimensions=dimensions_graph)
        self.graph_edge = spa.Buffer(dimensions=dimensions_graph, subdimensions=8)
        self.graph_next = spa.Buffer(dimensions=dimensions_graph, subdimensions=8)
        
        graph_edge = AutoAssociative(
            neurons_per_item*len(initial_edge_keys), max_capacity,
            initial_keys=initial_edge_keys, initial_values=initial_edge_values, use_all_encoders=True)
        nengo.Connection(self.graph_key.state.output, graph_edge.key)
        nengo.Connection(graph_edge.output, self.graph_edge.state.input)
        
        graph_next = AutoAssociative(
            neurons_per_item*len(initial_next_keys), max_capacity,
            initial_keys=initial_next_keys, initial_values=initial_next_values, use_all_encoders=True)
        nengo.Connection(self.graph_key.state.output, graph_next.key)
        nengo.Connection(graph_next.output, self.graph_next.state.input)

        # Create a queue using a Dict and interface with SPA blocks
        self.queue = AutoAssociative(
            neurons_per_item*num_vertices, AutoAssociative.calculate_max_capacity(dimensions_graph, 0.8), 
            dimensions_graph, dimensions_graph,
            pes_learning_rate=1e-3)
        self.queue_key = spa.DoubleLatch(dimensions=dimensions_graph)
        self.queue_value = spa.DoubleLatch(dimensions=dimensions_graph)
        self.queue_learning = spa.DoubleLatch(dimensions=1)
        self.queue_output = spa.Buffer(dimensions=dimensions_graph, subdimensions=8)
        self.queue_error = spa.Buffer(dimensions=dimensions_graph, subdimensions=8)
        self.queue_has_key = spa.Buffer(dimensions=1, subdimensions=1)
        nengo.Connection(self.queue_key.state.output, self.queue.key)
        nengo.Connection(self.queue_value.state.output, self.queue.value)
        nengo.Connection(self.queue_learning.state.output, self.queue.learning)
        nengo.Connection(self.queue.output, self.queue_output.state.input)
        nengo.Connection(self.queue.has_key, self.queue_has_key.state.input)
        nengo.Connection(self.queue.value_error, self.queue_error.state.input)

        # Variables
        self.state = spa.Memory(dimensions=dimensions_state)
        self.temp = spa.DoubleLatch(dimensions=dimensions_graph)
        self.queue_head = spa.DoubleLatch(dimensions=dimensions_graph)
        self.queue_tail = spa.DoubleLatch(dimensions=dimensions_graph)
        
        # Input
        self.v_0 = spa.Buffer(dimensions=dimensions_graph, subdimensions=8)
        
        # Control flow
        actions = self.control_dfa.actions
        for name, action in actions.items():
            print "[%s] %s" % (name, action)
        actions = spa.Actions(**actions)
        self.bg = spa.BasalGanglia(actions=actions, radius=3, n_neurons_per_ensemble=250)
        self.thal = spa.Thalamus(self.bg, channel_subdim=8)
        
        # Initialize starting vertex and state
        self.input_vertex = spa.Input(v_0=lambda t: v_0)
        self.input_state = spa.Input(state=lambda t: '%s*%s' % (5.0, self.control_dfa.initial) if t < 0.02 else '0')  # TODO
```

```python
adjacency_dict = {
    '0': ['1', '2'],
    '1': ['3', '0'],
    '2': ['0'],
    '3': ['1', '0', '2'],
}

adjacency_dict = {
    '0': ['0', '3'],
    '1': ['4'],
    '2': ['2'],
    '3': ['1', '0', '3', '2'],
    '4': ['2'],
}

#adjacency_dict = BFS.control_dfa.adjacency_dict

model = nengo.Network()
with model:
    bfs = BFS(adjacency_dict, '0', 16, 32)

    modules = ('v_0', 'state', 'graph_key', 'graph_key_latched', 
               'graph_edge', 'graph_next',
               'queue_head', 'queue_head_latched', 'queue_tail', 'queue_tail_latched', 
               'queue_key', 'queue_key_latched', 'queue_value', 'queue_value_latched', 
               'queue_output', 'queue_error', 'queue_has_key', 
               'queue_learning', 'queue_learning_latched',
               'temp', 'temp_latched')
    module_probes = OrderedDict([
        (module, nengo.Probe(bfs.get_module_output(module)[0], filter=0.01))
        for module in modules])
    actions_probe = nengo.Probe(bfs.thal.actions.output, filter=0.03)
    utilities_probe = nengo.Probe(bfs.bg.input, filter=0.03)
    bg_output_probe = nengo.Probe(bfs.bg.output, filter=0.03)
    
```

```python
sim = nengo.Simulator(model)
sim.run(20)
```

```python
vert = ['0', '3', '1', '2', '4'][::-1]
#vert = ['INIT', 'IF', 'NOTIN_Q', 'IN_Q', 'ADDED_Q', 'NEXT_ADJ', 'END_ADJ', 'POP_Q'][::-1]
duration = 19
t = sim.trange()
sl = (0.1 < t) & (t < duration)

def do_vertex_plot(module, label):
    data = module_data[module]
    vocab = bfs.get_module_output(module)[1]
    y = []
    
    window = 1 # numpy.ones(5000)/float(5000)
    pts = np.array([np.convolve(p, window, 'same') for p in np.dot(data, vocab.vectors.T).T]).T
    for p in pts:
        try:
            y.append(vert.index(vocab.keys[np.argmax(p)].split('_', 1)[-1]))
        except ValueError:
            y.append(y[-1])
    plt.plot(t[sl], np.asarray(y)[sl], label=label, linewidth=13, alpha=0.4)

plt.figure(figsize=(10, 2))
plt.title('Queue State for Toy Graph')
plt.xlabel('Time (s)')
plt.ylabel('Vertex')
plt.yticks(range(len(vert)), vert)
plt.ylim(-0.5, len(vert)+0.5)
do_vertex_plot('queue_head', 'Head')
do_vertex_plot('queue_tail', 'Tail')
plt.vlines(np.where(module_data['queue_learning'][sl] > 0.5)[0]*0.001, -0.5, len(vert)+0.5, alpha=0.01, color='orange', label='Learning')
plt.legend()
```
