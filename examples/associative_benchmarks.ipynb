```python
import numpy as np
np.random.seed(11)

from collections import Counter

import matplotlib.pyplot as plt

import nengo
from nengo.networks.associative import AutoAssociative
from nengo.utils.distributions import UniformHypersphere
from nengo.utils.ensemble import encoders
```

```python
from scipy.linalg import sqrtm, inv

def orthonormal(m):
    return m.dot(inv(sqrtm(m.T.dot(m)))).real
```

```python
n_neurons = 2500
d_value = 1
period = 0.13
delay = 0.11
pstc = 0.02

def iterate_periodically(l, period):
    return lambda t: l[(t/period) % len(l)]
```

```python
def gen_keys(d, n, extra=0.5, attempts=1000):
    c = AutoAssociative._calculate_intercept(d, 1.0 / (extra * n))
    keys = []
    for i in range(n):
        for _ in range(attempts):
            key = UniformHypersphere(d, surface=True).sample(1)[0]
            if len(keys) == 0 or np.asarray(keys).dot(key).max() < c:
                break
        else:
            raise ValueError("%d, %d" % (d, n))        
        keys.append(key)
    return np.asarray(keys)
```

```python
num_trials = 25
x = [10, 30, 50, 75, 100, 130, 160, 200, 250]
all_d = [8, 16, 32, 64, 128]

all_results = []
for d_key in all_d:
    results = []
    for num_items in x:
        trial_results = []
        for trial in range(num_trials):
            print d_key, num_items, trial
            
            keys = gen_keys(d_key, num_items)
            values = UniformHypersphere(d_value).sample(num_items)
            
            m = AutoAssociative(n_neurons, num_items, initial_keys=keys, initial_values=values)
            assert (keys.dot(keys.T) - 2*np.eye(num_items)).max() < m.intercept
            
            with m:
                key = nengo.Node(output=iterate_periodically(keys, period))            
                nengo.Connection(key, m.key, filter=None)
                p = nengo.Probe(m.output, filter=pstc)
            
            counter = Counter(tuple(e) for e in m.memory.encoders)
            counts = []
            for key in keys:
                counts.append(counter[tuple(key)])
            
            sim = nengo.Simulator(m)
            sim.run(period*num_items)
            t = sim.trange()
            
            rmse = 0
            errors = []
            for i in range(num_items):
                r = (i*period + delay < t) & (t < (i+1)*period)
                errors.append(np.mean((sim.data[p][r] - values[i])**2))
            trial_results.append((errors, counts))
        results.append(trial_results)
    all_results.append(results)
```

```python
all_y1 = []
all_y1err = []
all_y2 = []
all_y2err = []
all_y3 = []
all_y3err = []
correct_thresh = 0.02
for results in all_results:
    y1 = []
    y1err = []
    y2 = []
    y2err = []
    y3 = []
    y3err = []
    for trial_results in results:
        p1 = []
        p2 = []
        p3 = []
        for error, count in trial_results:
            p1.append(np.sqrt(np.mean(np.asarray(error))))
            p2.append(len(np.where(np.sqrt(np.asarray(error)) < correct_thresh)[0]) / float(len(error)))
            p3.append(np.mean(count))
        for y, yerr, p in zip((y1, y2, y3), (y1err, y2err, y3err), (p1, p2, p3)):
            p = np.asarray(p)
            mu = np.mean(p)
            std = np.sqrt(np.mean((p - mu)**2))
            y.append(mu)
            yerr.append(std)
    all_y1.append(y1)
    all_y1err.append(y1err)
    all_y2.append(y2)
    all_y2err.append(y2err)
    all_y3.append(y3)
    all_y3err.append(y3err)
```

```python
markers = ['o', 'v', 's', 'h', 'D', 'x']

plt.figure()
plt.title('Representational Accuracy of Values')
for d, y, yerr, marker in zip(all_d, all_y1, all_y1err, markers):
    plt.errorbar(x, y, yerr=yerr, label="d=%d" % d, marker=marker)
plt.xlim(0, 260)
plt.ylim(0)
plt.xlabel('Number of Associations')
plt.ylabel('RMSE')
plt.legend(loc='upper right')

plt.figure()
plt.title('Proportion of Values with RMSE < %s' % correct_thresh)
for d, y, yerr, marker in zip(all_d, all_y2, all_y2err, markers):
    plt.errorbar(x, y, yerr=yerr, label="d=%d" % d, marker=marker)
plt.xlim(0, 260)
plt.ylim(0.8, 1.1)
plt.xlabel('Number of Associations')
plt.ylabel('Percent Correct')
plt.legend(loc='upper right')

from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

fig = plt.figure()
ax = fig.add_subplot(111)
plt.title('Number of Neurons per Association')
x = np.asarray(x)
for d, y, yerr, marker in zip(all_d, all_y, all_yerr, markers):
    ax.errorbar(x, y, yerr=yerr, label="d=%d" % d, marker=marker)
yopt = n_neurons/x
ax.plot(x, yopt, label="Optimal", marker="*")
ax.set_xlim(0, 260)
ax.set_ylim(0, 260)
ax.set_xlabel('Number of Associations')
ax.set_ylabel('Number of Neurons')
ax.legend(loc='upper right')

sl = (x >= 130) & (x <= 160)

axins = zoomed_inset_axes(ax, 4, bbox_to_anchor=(190, 230), loc=9)

for d, y, yerr, marker in zip(all_d, all_y3, all_y3err, markers):
    axins.errorbar(x[sl], np.asarray(y)[sl], yerr=np.asarray(yerr)[sl], label="d=%d" % d, marker=marker)
axins.plot(x[sl], yopt[sl], label="Optimal", marker="*")

x1, x2, y1, y2 = 129, 161, 5, 25
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
axins.set_xticks([])

mark_inset(ax, axins, loc1=3, loc2=4, fc="none", ec="0.5")
```
