```python
import numpy as np
np.random.seed(11)

import nengo
from nengo.networks import DoubleLatch
from nengo.utils.distributions import UniformHypersphere
```

```python
speeds = np.arange(0.2, 1.51, 0.1)
N = [50]
num_trials = 25
dimensions = 16
sim_time = 1.0
offset = 0.1
th = 0.5
sample_offset = 100
sample_length = 300

all_data = []
for n in N:
    data = []
    for speed in speeds:
        print n, speed
        trials = []
        for trial in range(num_trials):
            input = UniformHypersphere(dimensions, surface=True).sample(1)[0]
            
            m = nengo.Network()
            with m:
                dli = DoubleLatch(dimensions, neurons_per_dimension=n, speed=speed, double_latch_threshold=0.95)
                
                input_node = nengo.Node(size_in=1, output=lambda t, x: input if (t > offset and x[0] < th) else np.zeros(dimensions))
                nengo.Connection(dli.latched, input_node)
                nengo.Connection(input_node, dli.input, filter=None)
                
                latch = nengo.Node(size_in=1, output=lambda t, x: int(t > offset))
                nengo.Connection(dli.latched, latch)
                nengo.Connection(latch, dli.latch)
                
                p_input = nengo.Probe(input_node, filter=0.005)
                p_latched = nengo.Probe(dli.latched, filter=0.005)
                p_output = nengo.Probe(dli.state.output, filter=0.005)
                p_error = nengo.Probe(dli.inhibited_error.output, filter=0.03)
                
            sim = nengo.Simulator(m)
            sim.run(sim_time)
            
            latched = sim.data[p_latched]
            output = sim.data[p_output]
            t = sim.trange()
            
            '''
            plt.figure()
            plt.plot(t, np.dot(sim.data[p_input], input.T), label="Input")
            plt.plot(t, latched, label="Latched")
            plt.plot(t, np.dot(output, input.T), label="Output")
            plt.plot(t, [np.dot(e, e) for e in sim.data[p_error]], label="Error")
            plt.legend()
            plt.show()
            '''
            
            sl = latched.flatten() >= th
            latched_time = t[sl][0]
            rmse = np.sqrt(np.mean((input - output[sl][sample_offset:sample_offset+sample_length])**2))
            
            trials.append((latched_time, rmse))
        data.append(trials)
    all_data.append(data)
```

```python
def input(t):
    return [1, -1, 0][int(t % 3)]

def latch(t):
    return [1, 1, 1, 0][int(t % 4)]

m = nengo.Network()
with m:
    dli = DoubleLatch(1, double_latch_threshold=0.99)
    
    input_node = nengo.Node(output=input)
    latch_node = nengo.Node(output=latch)
    nengo.Connection(input_node, dli.input)
    nengo.Connection(latch_node, dli.latch)
    
    p_input = nengo.Probe(input_node, filter=0.005)
    p_latch = nengo.Probe(latch_node, filter=0.005)
    p_output = nengo.Probe(dli.state.output, filter=0.03)
    p_latched = nengo.Probe(dli.latched, filter=0.03)
    
sim = nengo.Simulator(m)
sim.run(12)
```

```python
figure(figsize=(10, 3))
title('DLI Inputs')
xlabel('Time (s)')
ylabel('Value')
ylim(-1.1, 1.1)
plot(sim.trange(), sim.data[p_input], label="Input", c="green", linewidth=3, alpha=0.8)
plot(sim.trange(), sim.data[p_latch], label="Latch", c="red", linewidth=3, alpha=0.8)
legend(loc='lower right')

figure(figsize=(10, 3))
title('DLI Outputs')
xlabel('Time (s)')
ylabel('Value')
ylim(-1.1, 1.1)
plot(sim.trange(), sim.data[p_output], label="Output", c="green", linewidth=3, alpha=0.8)
plot(sim.trange(), sim.data[p_latched], label="Latched", c="red", linewidth=3, alpha=0.8)
legend(loc='lower right')
```
